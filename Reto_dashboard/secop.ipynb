## Cargar data

```python
import pandas as pd
import requests

# URL de la API del SECOP Integrado
api_url = "https://www.datos.gov.co/resource/rpmr-utcd.json?$limit=50000"

# Cargar datos
try:
    response = requests.get(api_url)
    response.raise_for_status()
    data = response.json()
    df_raw = pd.DataFrame(data)
    print(f"✅ Datos cargados: {df_raw.shape[0]} filas, {df_raw.shape[1]} columnas")
    
    # Ver primeras filas
    display(df_raw.head())

except Exception as e:
    print(f"❌ Error al cargar los datos: {e}")
    df_raw = pd.DataFrame()  # Para que df_raw exista aunque falle
```

```python
# Ver las primeras filas
df_raw.head()
```

```python
# Copia de trabajo
df = df_raw.copy()

# ------------------------------
# 1. Limpiar nombres de columnas (eliminar espacios y convertir a minúsculas)
# ------------------------------
df.columns = df.columns.str.strip().str.lower()

# ------------------------------
# 2. Convertir columnas de fecha a tipo datetime
# ------------------------------
for col in ['fecha_inicio_ejecuci_n', 'fecha_fin_ejecuci_n']:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# ------------------------------
# 3. Convertir columnas numéricas a tipo float (valor del contrato, documentos, NIT, etc.)
# ------------------------------
df['valor_contrato'] = pd.to_numeric(df['valor_contrato'], errors='coerce')
df['documento_proveedor'] = df['documento_proveedor'].str.replace(".", "", regex=False)
df['documento_proveedor'] = pd.to_numeric(df['documento_proveedor'], errors='coerce')

# ------------------------------
# 4. Estandarizar texto: quitar espacios, pasar a minúsculas, limpiar NaNs
# ------------------------------
texto_cols = df.select_dtypes(include='object').columns
for col in texto_cols:
    df[col] = df[col].fillna("").str.strip()

# ------------------------------
# 5. Eliminar duplicados si los hay
# ------------------------------
df.drop_duplicates(inplace=True)

# ------------------------------
# 6. Ver resumen limpio
# ------------------------------
print("✅ Datos limpiados:")
print(df.info())
display(df.head())
```

```python
#  limpiar y estandarizar los nombres de las columnas
import unidecode
df.columns = [unidecode.unidecode(col).lower().replace(" ", "_") for col in df.columns]
```

### ¿Qué departamentos del país presentan una mayor tasa de contratos por cada 1.000 habitantes?

```python
ruta_2005 = "C:/Users/natam/OneDrive/Documentos/Diplomado/Datos/Info_2005_2019.xlsx"
df_2005_2019 = pd.read_excel(ruta_2005)
ruta_2020 = "C:/Users/natam/OneDrive/Documentos/Diplomado/Datos/Info_2020_2035.xlsx"
df_2020_2035 = pd.read_excel(ruta_2020)
```

```python
print("📄 Columnas 2005–2019:", df_2005_2019.columns.tolist())
print("📄 Columnas 2020–2035:", df_2020_2035.columns.tolist())

# Ver algunas filas para confirmar estructura
df_2020_2035.head()
```

```python
# Unir ambos DataFrames
df_pob = pd.concat([df_2005_2019, df_2020_2035], ignore_index=True)

# Eliminar filas sin año
df_pob = df_pob[df_pob['AÑO'].notna()]

# Asegurar que 'AÑO' sea tipo entero
df_pob['AÑO'] = df_pob['AÑO'].astype(int)

# Filtrar solo "Total" por área geográfica
df_pob_total = df_pob[df_pob['ÁREA GEOGRÁFICA'].str.lower() == 'total']

# Seleccionar el año más reciente
año_reciente = df_pob_total['AÑO'].max()

# Filtrar datos solo del año más reciente
df_pob_ult = df_pob_total[df_pob_total['AÑO'] == año_reciente]

# Verificar los datos resultantes
print(f"📅 Año más reciente: {año_reciente}")
print(df_pob_ult.head())
```

```python
# Agrupar población total por departamento
df_pob_departamento = df_pob_ult.groupby('DPNOM', as_index=False)['Población'].sum()

# Renombrar columna para unir más fácilmente después
df_pob_departamento.rename(columns={'DPNOM': 'departamento_entidad', 'Población': 'poblacion_2035'}, inplace=True)

# Mostrar resultado
df_pob_departamento.head()
```

```python
# 1. Contar contratos por departamento
df_contratos_por_dep = df_raw.groupby('departamento_entidad', as_index=False).size()
df_contratos_por_dep.rename(columns={'size': 'num_contratos'}, inplace=True)

# 2. Unir con la población
df_tasa = pd.merge(df_contratos_por_dep, df_pob_departamento, on='departamento_entidad', how='left')

# 3. Calcular tasa de contratos por cada 1.000 habitantes
df_tasa['contratos_por_1000_hab'] = (df_tasa['num_contratos'] / df_tasa['poblacion_2035']) * 1000

# 4. Revisar los primeros resultados
df_tasa.sort_values(by='contratos_por_1000_hab', ascending=False).head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ordenar por tasa
df_top = df_tasa.sort_values(by='contratos_por_1000_hab', ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=df_top, x='contratos_por_1000_hab', y='departamento_entidad', palette='viridis')
plt.xlabel('Contratos por cada 1.000 habitantes')
plt.ylabel('Departamento')
plt.title('Top 10 departamentos con mayor tasa de contratación')
plt.tight_layout()
plt.show()
```

¿Se corresponde el volumen de contratación con la población o hay departamentos que reciben proporcionalmente más o menos contratos?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# --- 1. Cargar archivos de población ---
ruta_2005 = "C:/Users/natam/OneDrive/Documentos/Diplomado/Datos/Info_2005_2019.xlsx"
ruta_2020 = "C:/Users/natam/OneDrive/Documentos/Diplomado/Datos/Info_2020_2035.xlsx"

df_2005_2019 = pd.read_excel(ruta_2005)
df_2020_2035 = pd.read_excel(ruta_2020)

df_pob = pd.concat([df_2005_2019, df_2020_2035], ignore_index=True)

# --- 2. Filtrar solo "Total" por departamento ---
df_pob = df_pob[df_pob['ÁREA GEOGRÁFICA'].str.lower() == 'total']

# --- 3. Filtrar solo año 2035 ---
df_pob_2035 = df_pob[df_pob['AÑO'] == 2035]

# --- 4. Agrupar población por departamento ---
df_poblacion = df_pob_2035.groupby('DPNOM')['Población'].sum().reset_index()
df_poblacion.rename(columns={'DPNOM': 'departamento_entidad', 'Población': 'poblacion_2035'}, inplace=True)

# --- 5. Contar contratos por departamento ---
df_contratos_por_depto = df_raw.groupby('departamento_entidad').size().reset_index(name='num_contratos')

# --- 6. Unir contratos con población ---
df_contratos_pob = df_contratos_por_depto.merge(df_poblacion, on='departamento_entidad', how='inner')

# --- 7. Gráfico de dispersión ---
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_contratos_pob, x='poblacion_2035', y='num_contratos', hue='departamento_entidad')
plt.xlabel('Población estimada en 2035')
plt.ylabel('Número de contratos')
plt.title('Relación entre población y número de contratos por departamento')
plt.grid(True)
plt.legend([], [], frameon=False)  # Oculta leyenda
plt.show()

# --- 8. Correlación ---
corr, _ = pearsonr(df_contratos_pob['poblacion_2035'], df_contratos_pob['num_contratos'])
print(f'📈 Correlación de Pearson: {corr:.2f}')
```

El gráfico muestra una correlación positiva fuerte (r = 0.92) entre la población estimada en 2035 y el número de contratos por departamento en Colombia. Es decir, los departamentos con más población tienden a tener más contratos. Sin embargo, hay un posible outlier con valores muy altos, probablemente Bogotá o Antioquia. Esto sugiere que la asignación de contratos está, en general, alineada con el tamaño poblacional.

### ¿Cómo ha sido la evolución mensual del valor total contratado en los últimos años, según el tipo de contrato?

```python
# 1. Limpiar columna de tipo de contrato
df['tipo_de_contrato'] = df['tipo_de_contrato'].fillna("").str.strip().str.lower()

# 2. Crear columna año-mes para agrupación (usamos la fecha de inicio de ejecución)
df['anio_mes'] = df['fecha_inicio_ejecuci_n'].dt.to_period('M')

# 3. Agrupar por año-mes y tipo de contrato, sumando valor total contratado
df_evolucion = df.groupby(['anio_mes', 'tipo_de_contrato'])['valor_contrato'].sum().reset_index()

# Convertir 'anio_mes' a datetime para graficar fácilmente
df_evolucion['anio_mes'] = df_evolucion['anio_mes'].dt.to_timestamp()

# 4. Filtrar últimos años, por ejemplo desde 2018 en adelante
df_evolucion_reciente = df_evolucion[df_evolucion['anio_mes'].dt.year >= 2018]

# 5. Gráfica de evolución mensual
plt.figure(figsize=(14, 7))
sns.lineplot(data=df_evolucion_reciente, x='anio_mes', y='valor_contrato', hue='tipo_de_contrato', marker='o')

plt.title('Evolución mensual del valor total contratado por tipo de contrato (desde 2018)')
plt.xlabel('Fecha (Año-Mes)')
plt.ylabel('Valor total contratado')
plt.legend(title='Tipo de contrato', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Evolución mensual: Se observan fluctuaciones claras en el valor total contratado a lo largo del tiempo, con variaciones según el tipo de contrato.
Picos importantes: Hay aumentos marcados en ciertos meses, especialmente en contratos de obra y suministro, lo que podría coincidir con cierres fiscales o planes de inversión pública.
Caídas notables: También se evidencian caídas abruptas, posiblemente por cambios administrativos, restricciones presupuestales o eventos externos como la pandemia.
