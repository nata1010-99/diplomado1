```python
import pandas as pd
import requests
import time
import math
from requests.exceptions import ChunkedEncodingError

def descargar_secop_por_lote(offset, limit):
    url = f"https://www.datos.gov.co/resource/rpmr-utcd.json?$limit={limit}&$offset={offset}"
    headers = {"User-Agent": "Mozilla/5.0"}

    try:
        r = requests.get(url, headers=headers, timeout=60)
        if r.status_code != 200 or not r.text.startswith("["):
            print(f"⚠️ Offset {offset} falló. Código HTTP: {r.status_code}")
            return pd.DataFrame()
        data = r.json()
        if len(data) < 50:
            print(f"⚠️ Lote con offset {offset} tiene pocos registros ({len(data)}). Deteniendo.")
            return pd.DataFrame()
        df = pd.DataFrame(data)

    except ChunkedEncodingError as e:
        print(f"⚠️ Conexión rota en offset {offset}: {e}")
        time.sleep(5)
        return pd.DataFrame()
    except Exception as e:
        print(f"❌ Error inesperado en offset {offset}: {e}")
        time.sleep(5)
        return pd.DataFrame()

    for col in ["objeto_a_contratar", "objeto_del_proceso"]:
        if col in df.columns:
            df[col] = df[col].astype(str)\
                .str.replace("\n", " ", regex=False)\
                .str.replace("'", "", regex=False)\
                .str.strip()

    df = df.dropna(thresh=10)
    return df

# 🚀 Descarga con reintentos automáticos
total_registros = 19500000
num_lotes = 20
tamaño_lote = math.ceil(total_registros / num_lotes)

for i in range(num_lotes):
    offset = i * tamaño_lote
    print(f"📦 Descargando lote {i+1}/{num_lotes} (offset: {offset})")

    max_reintentos = 3
    reintentos = 0
    df_lote = pd.DataFrame()

    while reintentos < max_reintentos:
        df_lote = descargar_secop_por_lote(offset, tamaño_lote)
        if not df_lote.empty:
            break
        print(f"🔁 Reintentando offset {offset} ({reintentos+1}/{max_reintentos})")
        reintentos += 1
        time.sleep(5)

    if df_lote.empty:
        print(f"📴 Offset {offset} fallido tras {max_reintentos} intentos. Se detiene.")
        break

    nombre_archivo = f"secop_lote_{i+1}.csv"
    df_lote.to_csv(nombre_archivo, index=False)
    print(f"✅ Guardado: {nombre_archivo} con {df_lote.shape[0]} filas.")
    time.sleep(1)
```

```python
import pandas as pd
import requests
import time
from requests.exceptions import ChunkedEncodingError

def descargar_secop_por_lote(offset, limit):
    url = f"https://www.datos.gov.co/resource/rpmr-utcd.json?$limit={limit}&$offset={offset}"
    headers = {"User-Agent": "Mozilla/5.0"}

    try:
        r = requests.get(url, headers=headers, timeout=120)

        if r.status_code != 200:
            print(f"⚠️ Offset {offset} falló. Código HTTP: {r.status_code}")
            return pd.DataFrame()

        try:
            data = r.json()
        except Exception as e:
            print(f"❌ JSON inválido en offset {offset}: {e}")
            return pd.DataFrame()

        if not isinstance(data, list) or len(data) < 50:
            print(f"⚠️ Offset {offset} retornó pocos registros o estructura inválida.")
            return pd.DataFrame()

        df = pd.DataFrame(data)

    except ChunkedEncodingError as e:
        print(f"⚠️ Conexión rota en offset {offset}: {e}")
        time.sleep(10)
        return pd.DataFrame()
    except Exception as e:
        print(f"❌ Error inesperado en offset {offset}: {e}")
        time.sleep(10)
        return pd.DataFrame()

    for col in ["objeto_a_contratar", "objeto_del_proceso"]:
        if col in df.columns:
            df[col] = df[col].astype(str)\
                .str.replace("\n", " ", regex=False)\
                .str.replace("'", "", regex=False)\
                .str.strip()

    df = df.dropna(thresh=10)
    return df

# 🔁 Reanudar desde el lote 14
offset_inicial = 12675000
lote_inicial = 14
total_descargado = 12675000
limite_total = 20000000
tamaño_lote = 975000
max_reintentos = 3

# 🚀 Descarga lote a lote hasta completar 20 millones
while total_descargado < limite_total:
    print(f"📦 Descargando lote {lote_inicial} (offset: {offset_inicial})")

    reintentos = 0
    df_lote = pd.DataFrame()

    while reintentos < max_reintentos:
        df_lote = descargar_secop_por_lote(offset_inicial, tamaño_lote)
        if not df_lote.empty:
            break
        print(f"🔁 Reintentando offset {offset_inicial} ({reintentos+1}/{max_reintentos})")
        reintentos += 1
        time.sleep(10)

    if df_lote.empty:
        print(f"📴 Offset {offset_inicial} fallido tras {max_reintentos} intentos. Se detiene.")
        break

    registros_actuales = df_lote.shape[0]
    total_descargado += registros_actuales

    nombre_archivo = f"secop_lote_{lote_inicial}.csv"
    df_lote.to_csv(nombre_archivo, index=False)
    print(f"✅ Guardado: {nombre_archivo} con {registros_actuales} filas. Total acumulado: {total_descargado}")

    offset_inicial += tamaño_lote
    lote_inicial += 1
    time.sleep(10)

# 🎯 Final
print(f"\n🛑 Proceso terminado. Total descargado: {total_descargado} registros.")
```
